```python
features_x = ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']
x = new_df.iloc[:,:4].values
y = new_df.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.40, random_state = 50)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

##Predicting the values

```python
print(classifier.predict(sc.transform([[5.1,3.5,1.4,0.2]])))
print(classifier.predict(sc.transform([[6.3,2.5,4.9,1.5]])))
```

##Testing accuracy

```python
y_pred = classifier.predict(x_test)
result = classifier.score(x_test,y_test)
print(result)
```

##Confusion Matrix for Logistics Regression

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
```

## KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
classify = KNeighborsClassifier(n_neighbors = 4, metric = 'minkowski', p = 2)
classify.fit(x_train, y_train)
y_pred = classify.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

##Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot=True,cmap="PiYG")
```

## Support Vector Classifier

```python
from sklearn.svm import SVC
from sklearn import metrics
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

##Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot = True,cmap = "viridis")
```

##Support Vector Machine

```python
from sklearn import svm
from sklearn import metrics
clf = svm.SVC(kernel='rbf', random_state=0)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

##Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot = True, cmap = 'PuOr')
```

##DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
cl = DecisionTreeClassifier(criterion="entropy", max_depth=2,splitter='best')
cl = cl.fit(x_train,y_train)
y_pred = cl.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

##Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
